```python
# Load the ratings data
ratings_df = pd.read_csv(
    '/content/u.data',
    sep='\t',
    header=None,
    names=['user_id', 'item_id', 'rating', 'timestamp']
)

# Load the movie information data
movies_df = pd.read_csv(
    '/content/u.item',
    sep='|',
    header=None,
    encoding='latin-1', # Specify encoding if needed
    names=['item_id', 'movie title', 'release date', 'video release date', 'IMDb URL', 'unknown', 'Action', 'Adventure', 'Animation', "Children's", 'Comedy', 'Crime', 'Documentary', 'Drama', 'Fantasy', 'Film-Noir', 'Horror', 'Musical', 'Mystery', 'Romance', 'Sci-Fi', 'Thriller', 'War', 'Western']
)


# Display the first 5 rows and data types of ratings_df
print("Ratings DataFrame:")
display(ratings_df.head())
print("\nRatings DataFrame Info:")
ratings_df.info()

# Display the first 5 rows and data types of movies_df
print("\nMovies DataFrame:")
display(movies_df.head())
print("\nMovies DataFrame Info:")
movies_df.info()

# Print the number of unique users and movies in the ratings data
num_unique_users = ratings_df['user_id'].nunique()
num_unique_movies = ratings_df['item_id'].nunique()

print(f"\nNumber of unique users in ratings data: {num_unique_users}")
print(f"Number of unique movies in ratings data: {num_unique_movies}")
```

```python
from pandas import pivot_table

user_item_matrix = pivot_table(ratings_df, index='user_id', columns='item_id', values='rating', fill_value=0)

display(user_item_matrix.head())
```

```python
from sklearn.metrics.pairwise import cosine_similarity

user_similarity_matrix = cosine_similarity(user_item_matrix)

print("User Similarity Matrix:")
display(user_similarity_matrix)
```

```python
import numpy as np

def recommend_movies_user_based(user_id, user_item_matrix, user_similarity_matrix, movies_df, k=10):
    """
    Recommends movies to a user based on user-based collaborative filtering.

    Args:
        user_id: The ID of the target user.
        user_item_matrix: The user-item matrix.
        user_similarity_matrix: The user similarity matrix.
        movies_df: DataFrame containing movie information.
        k: The number of recommendations to return.

    Returns:
        A list of recommended movie titles.
    """
    # 1. Get the index of the target user in the similarity matrix
    user_index = user_item_matrix.index.get_loc(user_id)

    # 2. Get similarity scores for the target user and sort users
    user_similarities = user_similarity_matrix[user_index]
    similar_users_indices = user_similarities.argsort()[::-1]

    # 3. Exclude the target user themselves
    similar_users_indices = similar_users_indices[similar_users_indices != user_index]

    # 4. Identify movies rated by the target user
    target_user_rated_movies = user_item_matrix.loc[user_id][user_item_matrix.loc[user_id] > 0].index.tolist()

    # 5. Find movies rated by similar users
    recommended_movies = {}
    for similar_user_index in similar_users_indices:
        similar_user_id = user_item_matrix.index[similar_user_index]
        similar_user_rated_movies = user_item_matrix.loc[similar_user_id][user_item_matrix.loc[similar_user_id] > 0].index.tolist()

        # 6. Filter out movies already rated by the target user
        newly_recommended_movies = [
            movie_id for movie_id in similar_user_rated_movies
            if movie_id not in target_user_rated_movies
        ]

        # Add ratings from similar users for these movies
        for movie_id in newly_recommended_movies:
            if movie_id not in recommended_movies:
                recommended_movies[movie_id] = []
            rating = user_item_matrix.loc[similar_user_id, movie_id]
            recommended_movies[movie_id].append(rating)

    # 7. Sort remaining movies based on the average rating given by similar users
    averaged_ratings = {
        movie_id: np.mean(ratings) for movie_id, ratings in recommended_movies.items()
    }

    sorted_recommended_movies = sorted(
        averaged_ratings.items(),
        key=lambda item: item[1],
        reverse=True
    )

    # 8. Return the top K recommended movies
    top_k_recommendations = sorted_recommended_movies[:k]

    # Get movie titles
    recommended_movie_ids = [movie_id for movie_id, _ in top_k_recommendations]
    recommended_movie_titles = movies_df[movies_df['item_id'].isin(recommended_movie_ids)]['movie title'].tolist()

    return recommended_movie_titles

# Example usage: Recommend movies for user 1
user_id_to_recommend = 1
k_recommendations = 10
recommendations = recommend_movies_user_based(user_id_to_recommend, user_item_matrix, user_similarity_matrix, movies_df, k=k_recommendations)

print(f"Recommendations for User {user_id_to_recommend}:")
for movie_title in recommendations:
    print(f"- {movie_title}")
```

```python
def precision_at_k(recommended_items, actual_items, k):
    """
    Calculates Precision at K.

    Args:
        recommended_items: A list of recommended item IDs.
        actual_items: A list of item IDs that the user actually liked.
        k: The number of recommendations to consider.

    Returns:
        The precision at K score.
    """
    # Ensure k is not larger than the number of recommendations
    k = min(k, len(recommended_items))
    if k == 0:
        return 0.0

    # Count the number of recommended items that are in the actual items
    num_relevant_and_recommended = len(set(recommended_items[:k]).intersection(set(actual_items)))

    return num_relevant_and_recommended / k

# Iterate through a subset of users and calculate precision@k
user_ids_to_evaluate = user_item_matrix.index[::50] # Evaluate every 50th user
precision_scores = []

for user_id in user_ids_to_evaluate:
    # Generate recommendations for the user
    recommended_movie_titles = recommend_movies_user_based(
        user_id,
        user_item_matrix,
        user_similarity_matrix,
        movies_df,
        k=k_recommendations
    )

    # Get the item_ids for the recommended movie titles
    recommended_movie_ids = movies_df[movies_df['movie title'].isin(recommended_movie_titles)]['item_id'].tolist()

    # Identify movies the user actually rated positively (rating >= 4)
    actual_rated_movies = user_item_matrix.loc[user_id]
    positively_rated_movie_ids = actual_rated_movies[actual_rated_movies >= 4].index.tolist()

    # Calculate precision at K for the current user
    precision = precision_at_k(recommended_movie_ids, positively_rated_movie_ids, k_recommendations)
    precision_scores.append(precision)

# Calculate and print the average precision at K
average_precision_at_k = np.mean(precision_scores)
print(f"Average Precision@{k_recommendations} for User-Based Collaborative Filtering: {average_precision_at_k:.4f}")
```

```python
import numpy as np

def recommend_movies_item_based(user_id, user_item_matrix, item_similarity_matrix, movies_df, k=10):
    """
    Recommends movies to a user based on item-based collaborative filtering.

    Args:
        user_id: The ID of the target user.
        user_item_matrix: The user-item matrix.
        item_similarity_matrix: The item similarity matrix.
        movies_df: DataFrame containing movie information.
        k: The number of recommendations to return.

    Returns:
        A list of recommended movie titles.
    """
    # 1. Identify the movies the user has rated
    user_ratings = user_item_matrix.loc[user_id]
    rated_movie_ids = user_ratings[user_ratings > 0].index.tolist()

    # 2. Create a dictionary to store potential recommendations and their scores
    movie_scores = {}

    # 3. For each movie the user has rated
    for rated_movie_id in rated_movie_ids:
        # Find similar items using the item similarity matrix
        # Get the index of the rated movie in the similarity matrix
        rated_movie_index = user_item_matrix.columns.get_loc(rated_movie_id)
        item_similarities = item_similarity_matrix[rated_movie_index]

        # Get the item IDs corresponding to the similarity scores
        similar_item_indices = np.argsort(item_similarities)[::-1]

        # Iterate through similar items
        for similar_item_index in similar_item_indices:
            similar_item_id = user_item_matrix.columns[similar_item_index]
            similarity_score = item_similarities[similar_item_index]

            # Exclude the rated movie itself and movies already rated by the user
            if similar_item_id != rated_movie_id and similar_item_id not in rated_movie_ids:
                # If the movie is not already in movie_scores, initialize it
                if similar_item_id not in movie_scores:
                    movie_scores[similar_item_id] = 0.0

                # Accumulate weighted score (similarity * user's rating of the similar item)
                # Note: In item-based, we use the user's rating of the *rated* movie to weight the *similar* movie
                # The standard item-based formula is sum(similarity * rating) / sum(|similarity|)
                # However, a simpler weighted sum can also work for ranking
                # Let's use a simple weighted sum for now: similarity * user's rating of the rated item
                user_rating_of_rated_movie = user_ratings.loc[rated_movie_id]
                movie_scores[similar_item_id] += similarity_score * user_rating_of_rated_movie


    # 4. Rank the unrated movies based on the accumulated scores
    sorted_recommended_movies = sorted(
        movie_scores.items(),
        key=lambda item: item[1],
        reverse=True
    )

    # 5. Return the top k recommended movie titles
    top_k_recommendations = sorted_recommended_movies[:k]

    # Get movie titles
    recommended_movie_ids = [movie_id for movie_id, _ in top_k_recommendations]
    recommended_movie_titles = movies_df[movies_df['item_id'].isin(recommended_movie_ids)]['movie title'].tolist()

    return recommended_movie_titles

# Example usage: Recommend movies for user 1 using item-based filtering
user_id_to_recommend = 1
k_recommendations = 10
recommendations_item_based = recommend_movies_item_based(user_id_to_recommend, user_item_matrix, item_similarity_matrix, movies_df, k=k_recommendations)

print(f"Item-Based Recommendations for User {user_id_to_recommend}:")
for movie_title in recommendations_item_based:
    print(f"- {movie_title}")
```
